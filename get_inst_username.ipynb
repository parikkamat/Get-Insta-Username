# **Code to Get Insta User Name to CSV**

```python
import re
import pandas as pd # For creating dataframes of extracted information
from google.colab import drive
drive.mount('/content/gdrive')
```

# **Extract_Username extracts the username from url**

```python
def extract_username(url):
    instaa = []
    beg, sep, end = url.partition('.com/')
    instaa = end.strip('/')


    return instaa
```

# **Reading Text File**

```python
insta_doc = open("/content/gdrive/My Drive/Insta_username/insta.txt", "r")
data = insta_doc.read()
print(data)
```

```python
fragmented_text_stream = re.split("\n", data)

print(fragmented_text_stream)
```

```python
fragmented_dataFrames = pd.DataFrame(fragmented_text_stream, columns =['User_ID'])

fragmented_dataFrames.head(3)
```

```python
insta_list = []

for i in (fragmented_dataFrames["User_ID"]):
  insta_list.append(extract_username(i))
```

```python
insta_list
```

# **Store to Dataframe**

```python
upsated_insta_id = pd.DataFrame(insta_list, columns =['new_User_ID'])

upsated_insta_id
```

# **Save to CSV File**

```python
upsated_insta_id.to_csv("/content/gdrive/My Drive/Insta_username/username-insta.csv",index=False)
```
